```python
new_columns = {
    'work_year': 'ano_trabalho',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'USD',
    'employee_residence': 'residencia,
    'remote_ratio': 'remoto',
    'company_location': 'localizacao_empresa',
    'company_size': 'porte_empresa'
}
df.rename(columns=new_columns, inplace=True)
print("As colunas do DataFrame foram traduzidas:")
print(df.columns)
```

```python
df["nivel_experiencia"].value_counts()
```

```python
df["tipo_emprego"].value_counts()
```

```python
df['tipo_emprego'] = df['tipo_emprego'].replace({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Meio Período',
    'FL': 'Freelancer'
})
print(df['tipo_emprego'].value_counts())
```

```python
df['nivel_experiencia'] = df['nivel_experiencia'].replace({
    'SE': 'Sênior',
    'MI': 'Médio',
    'EN': 'Iniciante',
    'EX': 'Executivo'
})
print(df['nivel_experiencia'].value_counts())
```
